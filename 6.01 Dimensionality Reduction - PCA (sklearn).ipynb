### Import library

```python
import pandas as pd                   
import matplotlib.pyplot as plt        
import seaborn as sns           
from sklearn.decomposition import PCA  
```

### Prepare Data

```python
train_features = pd.read_csv('train_features.csv')
train_labels = pd.read_csv('train_labels.csv')
train_labels = train_labels.Survived

test_features = pd.read_csv('test_features.csv')
test_labels = pd.read_csv('test_labels.csv')
test_labels = test_labels.Survived
test_features
```

### Train Model

```python
pca = PCA()

train_features_tf = pca.fit_transform(train_features)

sns.scatterplot(
    x=train_features_tf[:, 0], 
    y=train_features_tf[:, 1],
    style=train_labels,
    hue=train_labels,
    palette=['red', 'green'])
```

### Test Model

```python
test_features_tf = pca.fit_transform(test_features)

sns.scatterplot(
    x=test_features_tf[:, 0], 
    y=test_features_tf[:, 1],
    style=test_labels,
    hue=test_labels,
    palette=['red', 'green'])
```
