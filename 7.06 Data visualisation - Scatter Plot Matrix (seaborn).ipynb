### Import Library

```python
import pandas as pd     
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

### Prepare Data

```python
test_features = pd.read_csv('test_features.csv')
test_labels = pd.read_csv('test_labels.csv')
test_features = test_features[['Age','Fare']]

Survived_mapDict={
    1:'Survived',
    0:'Not Survived'
}

y=test_labels['Survived'].map(Survived_mapDict)
test_features = pd.concat([test_features, y], axis=1)
test_features.head(10)
```

### Plot Scatter Plot Matrix

```python
plt.rc('font',family = 'Times New Roman',  size=22)
sns.pairplot(test_features, hue='Survived')
```

```python
g = sns.PairGrid(
    test_features, 
    vars=["Age", "Fare"], 
    hue="Survived", 
    diag_sharey=False, 
    palette=["red", "green"])

g.map_diag(plt.hist)

g.map_upper(sns.kdeplot)

g.map_lower(sns.scatterplot)

g.add_legend()
```
