### Import library

```python
import pandas as pd                   
import matplotlib.pyplot as plt        
import seaborn as sns           
from sklearn.manifold import TSNE
```

### Prepare Data

```python
train_features = pd.read_csv('train_features.csv')
train_labels = pd.read_csv('train_labels.csv')
train_labels = train_labels.Survived
```

### Test Model

#### Perplexity = 1

```python
tsne = TSNE(
    n_components=2,
    perplexity=1,
    random_state=1)

train_features_tf = tsne.fit_transform(train_features)

sns.scatterplot(
    x=train_features_tf[:, 0], 
    y=train_features_tf[:, 1],
    style=train_labels,
    hue=train_labels,
    palette=['red', 'green'])
```
