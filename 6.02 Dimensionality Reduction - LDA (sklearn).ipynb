### Import library

```python
import pandas as pd                   
import matplotlib.pyplot as plt        
import seaborn as sns           
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

### Prepare Data

```python
train_features = pd.read_csv('train_features.csv')
train_labels = pd.read_csv('train_labels.csv')
train_labels = train_labels.Survived

test_features = pd.read_csv('test_features.csv')
test_labels = pd.read_csv('test_labels.csv')
test_labels = test_labels.Survived
test_features
```

### Train Model

```python
lda = LinearDiscriminantAnalysis()

lda.fit(train_features, train_labels)

train_features_tf = lda.transform(train_features)
```

```python
sns.scatterplot(
    x=train_features_tf[:, 0], 
    y=train_features_tf[:, 0],
    style=train_labels,
    hue=train_labels,
    palette=['red', 'green'])
```

### Test Model

```python
test_features_tf = lda.transform(test_features)

sns.scatterplot(
    x=test_features_tf[:, 0], 
    y=test_features_tf[:, 0],
    style=test_labels,
    hue=test_labels,
    palette=['red', 'green'])
```
