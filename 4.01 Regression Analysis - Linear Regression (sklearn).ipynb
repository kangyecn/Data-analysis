### Import library

```python
import pandas as pd                   
import matplotlib.pyplot as plt        
import seaborn as sns           
from sklearn.linear_model import LinearRegression
```

### Prepare Data

```python
train_features = pd.read_csv('train_features.csv')
train_labels = pd.read_csv('train_labels.csv')
train_labels = train_labels.Survived

test_features = pd.read_csv('test_features.csv')
test_labels = pd.read_csv('test_labels.csv')
test_labels = test_labels.Survived
test_features
```

### Train model

```python
lr = LinearRegression().fit(train_features[['Age']].values, train_features['Fare'].values)
print(
    'Score on training data: ' 
    + str("{:.2%}".format(lr.score(train_features[['Age']].values, train_features['Fare'].values))))
```

```python
lr.intercept_, lr.coef_
```

### Test model

```python
print(
    'Score on training data: ' 
    + str("{:.2%}".format(lr.score(test_features[['Age']].values, test_features['Fare'].values))))
```
