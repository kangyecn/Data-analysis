### Import Library

```python
import pandas as pd                               
import matplotlib.pyplot as plt 
```

### Prepare Data

```python
test_features = pd.read_csv('test_features.csv')
test_features = test_features[['Age', 'Fare']]
test_features = test_features.sample(n=5, random_state=1)
test_features
```

### Plot Bar Chart

```python
plt.rc('font', family = 'Times New Roman',  size=22)
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(1, 1, 1)
test_features.plot.bar(stacked=True, ax=ax1)
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)
ax1.legend(bbox_to_anchor=(0.7, -0.15), ncol=3, frameon=True, title='Feature')
```

```python
plt.rc('font', family = 'Times New Roman',  size=22)
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(1, 1, 1)
test_features.plot.bar(stacked=False, ax=ax1)
for tick in ax1.get_xticklabels():
    tick.set_rotation(45)
ax1.legend(bbox_to_anchor=(0.7, -0.15), ncol=3, frameon=True, title='Feature')
```
